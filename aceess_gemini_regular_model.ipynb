```python
## call the model -- invokication -- is there any to call(invoke) the llm to generate the output
## 3 types of the invoaktion
## invoke() #when ever we want to ask send a signle query  give the output after performing the concatention
## stream() # when ever we want to send a single query  # the token by token it will give itrate object
## batch() # when ever  to want to ask multiple query at a time
```

```python
rmodel = GoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
```

```python
for tokens in rmodel.stream('what is AI'): #itrators
  print(tokens)
```

```python
answer = rmodel.batch(['what is the full from of dl','what is full from of AI'])
```

```python
answer[1]
```

```python
answer[0]
```

```python
answer = []
for prompt in (['what is the full from of dl','what is full from of AI']):
  answer.append(rmodel.invoke(prompt))
```

```python
brain = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
message=[("system","you are having a 20 plus year experience in python programinng")]

while True:
  user = input("type the user messgae")

  message.append(('user',user))

  ai = brain.invoke(message)

  message.append(('ai',ai.content))

  print("ai {}".format(ai.content))

  if user.lower() == "bye":
    break
```
