```python
import os
from langchain_google_genai import GoogleGenerativeAI,ChatGoogleGenerativeAI
```

```python
from google.colab import userdata
ge = userdata.get('gemini')
```

```python
os.environ["GOOGLE_API_KEY"] = ge
```

```python
model = GoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
pt = PromptTemplate(template="you having a {exp} years of the experience you are going to answer if the query will be based on {topic} now answer based on the {q}",
              input_variables=["topic","q"],partial_variables={'exp':10})
```

```python
pt.format(topic="opencv",q="what are the important topic in opencv")
```

```python
response = model.invoke(pt.format(topic="opencv",q="what are the important topic in opencv"))
```

```python
print(response)
```

```python
## how to create a chatprompt template -- we want to create a list of message -- input to chat model
```

```python
from langchain_core.prompts import HumanMessagePromptTemplate,SystemMessagePromptTemplate,AIMessagePromptTemplate,ChatPromptTemplate
```

```python
SystemMessagePromptTemplate.from_template("you haveing a {exp} experience bases on the experience answer the query")
```

```python
HumanMessagePromptTemplate.from_template("expain me about {topic} in {x} sentences")
```

```python
message = [SystemMessagePromptTemplate.from_template("you haveing a {exp} experience bases on the experience answer the query"),HumanMessagePromptTemplate.from_template("expain me about {topic} in {x} sentences")]
```

```python
ChatPromptTemplate.from_messages(message)
```

```python
cpt = ChatPromptTemplate.from_messages(message)
```

```python
prompt = cpt.format(exp=3,topic='ml',x=10)
```

```python
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
response = model.invoke(prompt).content
```
